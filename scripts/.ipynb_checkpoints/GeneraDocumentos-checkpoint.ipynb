```python
import pandas as pd  # también instalar pip install openpyxl
import re
from docxtpl import DocxTemplate
from datetime import datetime
#from googletrans import Translator
import win32com.client as win32
import os
from collections import Counter
```

```python
PATH_TEMPLATES = '../Templates/'
PATH_PLANTILLAS = '{}Plantillas/'.format(PATH_TEMPLATES)
PATH_OUTPUT = '../Output/'
PATH_FILES='../Files/'
PATH_SHELL='.'
DOC_MC = 'MC_BI.docx'
DOC_MO = 'MO_BI.docx'
DOC_SWP = 'SWP_BI.docx'
EXCEL_SWP = 'SWP_BI.xlsx'
DOC_COPR = 'COPR.docx'

ICON_EXCEL = os.path.abspath('../icoExcel.ico')
ICON_PLANO= os.path.abspath('../icoPlano.ico')
```

```python
def changeKindTemplate(tipoTemplate):
    tiposTemplate = [
        {
            'tipo': 'SPTdt',
            'description': 'Ficha Stored Procedure Teradata',
            'codeRegister': 'python Frame_LoadSP.py',
            'codeExecute': 'python Frame_EjecutaSP.py'
        },
        {
            'tipo': 'SPNz',
            'description': 'Ficha Stored Procedure Netezza',
            'codeRegister': 'python Frame_LoadSP.py',
            'codeExecute': 'python Frame_EjecutaSP.py'
        },
        {
            'tipo': 'ExTdt',
            'description': 'Ficha de Extracción de data desde Teradata',
            'codeRegister': 'python Frame_Load_Extract.py',
            'codeExecute': 'python Frame_Extractor.py'
        },
        {
            'tipo': 'ExNz',
            'description': 'Ficha de Extracción de data desde Netezza',
            'codeRegister': 'python Frame_Load_Extract.py',
            'codeExecute': 'python Frame_Extractor.py'
        },
        {
            'tipo': 'NzLoad',
            'description': 'Ficha de Carga de data a Netezza',
            'codeRegister': 'python Frame_Load_Excel_to_TDT.py',
            'codeExecute': 'python Frame_NZLoad.py'
        },
        {
            'tipo': 'FLoad',
            'description': 'Ficha de Carga de data a Teradata',
            'codeRegister': 'python Frame_Load_Excel_to_TDT.py',
            'codeExecute': 'python Frame_FLoad_TDT.py'
        },
        {
            'tipo': 'RNNz',
            'description': 'Ficha de Carga de data a Netezza',
            'codeRegister': 'python Frame_Load_Excel_to_TDT.py',
            'codeExecute': 'python Frame_TransformaRN_NZ.py'
        },
        {
            'tipo': 'Schd',
            'description': 'Ficha de automatización',
            'codeRegister': 'python Frame_LoadSchedule.py',
            'codeExecute': 'python Frame_EjecutaSchd.py'
        },
        {
            'tipo': 'GRNz',
            'description': 'Ficha de generación de reporte en Netezza',
            'codeRegister': 'python Frame_Load_Reporte_NZ.py',
            'codeExecute': 'python Frame_Genera_Reporte_NZ.py'
        }
    ]
    dicFind=[kind for kind in tiposTemplate if kind['tipo']==tipoTemplate]
    dicKind=dicFind[0] if dicFind else dict()
    #proceso= tiposTemplate.get(tipoTemplate,'Tipo de template no identificado')
    return dicKind
```

```python
def changeKindScript(kindObject):
    tipoScript={
        'T':'Table',
        'SP':'Stored Procedure',
        'VW':'View',
        'F':'Function'
    }
    proceso=tipoScript.get(kindObject,'Tipo de script no identificado')
    return proceso
```

```python
def readExcel(fileName):
    try:
        fileSchedule = os.path.abspath("{}{}".format(PATH_TEMPLATES, fileName))
        pdSchedule = pd.read_excel(fileSchedule, sheet_name='Schedule', skiprows=1)
        pdSchedule = pdSchedule.rename(columns={
                                       'Nombre Layout': 'layout', 'Tipo Schedule': 'tipoTemplate', 'Archivo Template': 'nameTemplate'})
        pdSelSchedule = pdSchedule.iloc[:, [1, 2, 14]].dropna()
        return pdSelSchedule
    except FileNotFoundError:
        return 'Archivo no encontrado'
    except Exception as e:
        return f"Error al leer el archivo: {e}"
```

```python
def correCaracter(filename):
    invalid_chars_pattern = r'[<>:"/\\|?*]'
    cadCorregida = re.sub(invalid_chars_pattern, '_', filename)
    return cadCorregida
```

```python
def pathFourDir(path):
    componentes_ruta = os.path.abspath(path).split(os.sep)
    lastFourDir = componentes_ruta[-4:]
    ruta_modificada = '/' + '/'.join(lastFourDir)+'/'
    return ruta_modificada
```

```python
def createRollbacks(dfData,numberFeat,create_roll,rutaFiles=None):
    dfRollBack=[]
    if create_roll: 
        esquemas={dfRow['esquema'] for dfRow in dfData}
        for e in esquemas:
            rollbacks= ['------ROLLBACKS PARA EL ESQUEMA {}--------'.format(e)]
            
            nameArchivo='ROLLBACK {}.sql'.format(e)
            pathRoll=("{}FEAT_{}/{}").format(PATH_FILES,numberFeat,nameArchivo)
            for row in dfData:
                if row['esquema']==e:
                    kind=''
                    if row['kindObject']=='Stored Procedure':
                        kind='PROCEDURE'
                    else:
                        kind=row['kindObject'].upper()
                
                    codeDrop='DROP {} {};'.format(kind,row['nameScript'][:row['nameScript'].rfind('_')].upper())
                    rollbacks.append(codeDrop)

            """
            Luego de haber generado el código del ROLLBACK se crean los archivos y se llena el diccionario a devolver
            """
            with open(pathRoll,'w') as file:
                for r in rollbacks:
                    file.write('{}\n'.format(r))
            dfRoll=dict(esquema=e,pathRoll=pathRoll)
            dfRollBack.append(dfRoll)
    else:
        archivos=os.listdir(rutaFiles)
        for s in archivos:
            if s.startswith('ROLLBACK'):
                
                dfRollBack.append(s)        
    
    return dfRollBack
        
            
```

```python
def main(pdProceso, create_rollbacks):

    numberFeat = 123
    regu = 231
    description = correCaracter('Canastas D –Formato 2B LA/LC')
    fechaUpdate = datetime.today().strftime("%Y-%m-%d")
    month = datetime.today().strftime('%B')
    monthCreated = 'Junio' #Translator().translate(month, src='en', dest='es').text
    yearCreated = datetime.today().strftime('%Y')

    
    schemaFiles = pdProceso.apply(lambda row: row.to_dict(), axis=1).tolist()

    #Eliminar el nzload y transformador dado a que tienen el mismo layout y en la documentación solo se coloca uno
    layout_nz=[layout for layout in schemaFiles if layout['tipoTemplate'] in ('RNNz','NzLoad')]
    layout_count_nz=Counter(layout['layout'] for layout in layout_nz)
    layout_duplicados={layout for layout,count in layout_count_nz.items() if count>1}
    nzDelete='RNNz'
    schemaFiles=[template for template in schemaFiles if not (template['layout'] in layout_duplicados and template['tipoTemplate']==nzDelete)]


    #Llenado de nombres de archivos
    for diccionario in schemaFiles:
        updateNameFile = "{}{}_ExcelP".format(
            PATH_TEMPLATES, diccionario['nameTemplate'])
        diccionario['fileAdjunto'] = updateNameFile
        dicKind=changeKindTemplate(diccionario['tipoTemplate'])
        diccionario['tipoTemplate']= "{}:\n{}".format(dicKind['description'],diccionario['layout'])
        diccionario['desTemplate']="{} {}".format(dicKind['description'],diccionario['layout'])
        diccionario['codeRegister']="{} {}".format(dicKind['codeRegister'],diccionario['nameTemplate'])
        diccionario['updatePathFile']="{}{}".format(pathFourDir(PATH_TEMPLATES), diccionario['nameTemplate'])
        diccionario['code']=dicKind  
    
    #Clasificación de querys según tipo
    pathScript=("{}FEAT_{}/").format(PATH_FILES,numberFeat)
    rutaFile=os.path.abspath(pathScript)
    archivos=os.listdir(rutaFile)
    dicFiles=[]
    
    for script in archivos:
        print(script)
        if not script.startswith('ROLLBACK'):
            file=dict()
            file['esquema']=script[0:script.find('.')]
            file['kindObject']=changeKindScript(script[script.rfind('_')+1:script.rfind('.')])
            updateScript='{}{}_script'.format(pathScript,script)
            file['fileAdjunto']=updateScript
            file['nameScript']=script
            dicFiles.append(file)
    
    dicFiles=sorted(dicFiles,key=lambda x:x['esquema'])


    #Creacción y obtención de rollbacks
    if create_rollbacks:
        dfRollback=createRollbacks(dicFiles,numberFeat,create_rollbacks)

    #Empieza la renderización 
    docPlantilla = os.path.abspath("{}{}".format(PATH_PLANTILLAS, DOC_MC))
    doc = DocxTemplate(docPlantilla)

    valores = {
        'numberFeat': numberFeat,
        'numberRegu': regu,
        'description': description,
        'dateDoc': fechaUpdate,
        'monthCreated': monthCreated,
        'yearCreated': yearCreated,
        'registros': schemaFiles,
        'scripts':dicFiles,
        'rollbacks':dfRollback
    }

    finalFile = os.path.abspath(
        '{}FEAT-{}_REGU-{}_MC-BI {}.docx'.format(PATH_OUTPUT, numberFeat, regu, description))

    doc.render(valores)
    doc.save(finalFile)

    
    word_app = win32.gencache.EnsureDispatch("Word.Application")
    word_app.Visible = False
    doc_word = word_app.Documents.Open(finalFile)

    foundMarFile= False
    try:
        for paragraph in doc_word.Paragraphs:
            for diccionario in schemaFiles:
                valorFind= diccionario['fileAdjunto']
                
                if valorFind in paragraph.Range.Text:
                    print('Encontrado:{}'.format(diccionario['fileAdjunto'].replace('_ExcelP','')))
                    fileAdjunto=os.path.abspath(diccionario['fileAdjunto'].replace('_ExcelP',''))
                    paragraph.Range.Text = paragraph.Range.Text.replace(valorFind, '')
                    ole_shape = paragraph.Range.InlineShapes.AddOLEObject(
                        ClassType="Excel.Sheet.12",
                        FileName=fileAdjunto,
                        LinkToFile=False,
                        DisplayAsIcon=True,
                        IconFileName=ICON_EXCEL,
                        IconIndex=0,    
                        IconLabel=diccionario['nameTemplate']
                    )
                    break
            for script in dicFiles:
                valorFind=script['fileAdjunto']
                
                if valorFind in paragraph.Range.Text:
                    nameScript=valorFind.replace('_script','')
                    print('Encontrado:{}'.format(nameScript))
                    scriptAdjunto=os.path.abspath(nameScript)
                    paragraph.Range.Text=paragraph.Range.Text.replace(valorFind, '')
                    ole_shape = paragraph.Range.InlineShapes.AddOLEObject(
                        ClassType="Text.Document",
                        FileName=scriptAdjunto,
                        LinkToFile=False,
                        DisplayAsIcon=True,
                        IconFileName=ICON_PLANO,
                        IconIndex=0,    
                        IconLabel=script['nameScript']
                    )
                    break
                    
            for roll in dfRollback:
                valor_find=roll['pathRoll']
                
                if valor_find in paragraph.Range.Text:
                    print("Encontrado:{}".format(valor_find))
                    script_adjunto=os.path.abspath(valor_find)
                    paragraph.Range.Text=paragraph.Range.Text.replace(valor_find,'')
                    ole_shape=paragraph.Range.InlineShapes.AddOLEObject(
                        ClassType="Text.Document",
                        FileName=script_adjunto,
                        LinkToFile=False,
                        DisplayAsIcon=True,
                        IconFileName=ICON_PLANO,
                        IconIndex=0,    
                        IconLabel=valor_find[valor_find.rfind("/")+1:]
                    )
                    break
                    
            
    except Exception as e:
        print(f"Error al acceder a los párrafos del documento: {e}")
        doc_word.Close()
        word_app.Quit()
        #os.remove(temp_doc_path)
        raise
    

    print('pin36')
    doc_word.Close()
    word_app.Quit()
```

```python
if __name__ == "__main__":
    print('prueba')
    tabla = readExcel('Sched_PLT_CHRN_FJ.xlsx')
    print(tabla)
    main(tabla,1)
```
